# nrow

```python
import pandas as pd

# Read the CSV file with the first 20 rows
df = pd.read_csv('outputa.csv', nrows=20)

# Define the function to rename names
def rename_name(name):
    if len(name) > 11:
        return name[:13]  # Slice the name to 19 characters
    else:
        return name

# Apply the function to the 'Name' column
df['Name'] = df['Name'].apply(rename_name)

# Display the DataFrame with modified 'Name' column
print(df)
```
